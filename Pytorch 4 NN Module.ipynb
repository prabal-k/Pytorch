## Model Architecture

single neuron(i.e perceptron) with 5 input features , which applies sigmoid fxn and generates a single output


```python
#----> create the model class

import torch
import torch.nn as nn

class perceptron_model(nn.Module):   # Must inherit the nn.Module class inorder to use "nn module"

    # constructor
    def __init__(self,num_features):

        #----> invoking the parent class(nn.module) constructor using super keyword

        super().__init__()

        #----> Creating neural network (i.e perceptron)
        self.linear = nn.Linear(num_features,1)  #Accepts 2 paramaters , (numer_of_input_features , number_of_outputs)

        #----> Creating activation function
        self.sigmoid = nn.Sigmoid()

    #----> forward pass:
    def forward(self,features):
      out = self.linear(features)  #passing the features to the linear network (i.e calculates z= wx+b)
      out = self.sigmoid(out)
      return out






```

## Creating a fake *dataset*

```python
#----> create a dataset
feature = torch.rand(10,5) # dataset with 10 rows and 5 columns

# ----> Create a model
model = perceptron_model(feature.shape[1])

# ----> call model for forward pass
model.forward(feature)  # or better approach to call the forward function is : model(feature)
```

### passed 10 rows of inputs and we got 10 correspodning output i.e **y_pred**

```python
# show model weights
model.linear.weight
```

```python
# show model bias
model.linear.bias
```
