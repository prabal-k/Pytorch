# Example- 3

# Single Perceptron (scalar inputs)

Training a perceptron : where x is cgpa , y is pass/fail



```python
x = torch.tensor(6.7) #Input cgpa = 6.7
y = torch.tensor(0.0) #Output =0 ,i.e fail

# here requires_grad is false , beacue we are not calculating derivatives w.r.t x or y
```

```python
# Initial weights and biases
w = torch.tensor(1.0,requires_grad=True) #Weight
b = torch.tensor(0.0,requires_grad=True) #Bias
```

```python
w,b
```

### Forward Propogation

```python
z = w*x+b
y_pred = torch.sigmoid(z)
```

```python
z,y_pred
```

## Calculate loss

```python
# Binary Cross-Entropy Loss for scalar
def binary_cross_entropy_loss(prediction, target):
    epsilon = 1e-8  # To prevent log(0)
    prediction = torch.clamp(prediction, epsilon, 1 - epsilon)
    return -(target * torch.log(prediction) + (1 - target) * torch.log(1 - prediction))
```

```python
loss = binary_cross_entropy_loss(y_pred,y)
loss
```

```python
loss.backward()
```

```python
print(w.grad)
print(b.grad)
```

# Disable gradient tracking

-> If derivatives is not required to be calculated
(in case like: When we train the model the gradients need to be tracked but once training is completed , during making prediction , gradient tracking is not required)

```python
x3 = torch.tensor(2.0,requires_grad=True) #  tensor
x3
```

```python
y3 = x3 **2
```

```python
y3.backward()
```

```python
x3.grad
```

```python
# Now no need to track the gradient
```

```python
# Option 1:
x3.requires_grad_(False)
```
