# Ways of Creating Tensors

```python
# Using empty :
#-> Creates a tensor of specified shape in the meomry and returns the pre-existing values in that memory space
torch.empty(2,3)
```

```python
# using zeros
#-> Creates a tensor of specified shape and initializes with all zero values
torch.zeros(2,3)
```

```python
# using ones
#-> Creates a tensor of specified shape and initializes with all ones values
torch.ones(3,3)
```

```python
# Using rand
#-> Creates a tensor of specified shape and initializes with all random values each time/runned with different values
torch.rand(3,4)
```

```python
# Use of manual_seed
torch.manual_seed(10)
torch.rand(3,4)
#-> Creates a tensor of specified shape and initializes with all random values but each time/run with same values (for reproducability and debugging)
```

```python
# Using tensor
#-> Creates a custom tensors of specified shape where values can we manually defined
torch.tensor([[1,2,3],
              [3,5,7]])
```

```python
# Uing full
#-> Creates a custom tensors of with specified shape(3,4) where all values are fulled with specified value(5)
torch.full((3,4),5)
```

# To copy the shape of the existing tensor

```python
#-> Copies the shape of tensor 'a' i.e (2,3) and creates a new tensor of (2,3) but with different values
torch.empty_like(a)
```

```python
#-> Copies the shape of tensor 'a' i.e (2,3) and creates a new tensor of (2,3) but with all zeroes value
torch.zeros_like(a)
```

```python
#-> Copies the shape of tensor 'a' i.e (2,3) and creates a new tensor of (2,3) but with all ones value
torch.ones_like(a)
```

```python
torch.rand_like(a)
```

###### Problem with above code is that out tensor 'a' if of int datatype but rand() fxn creates a tensor with floatdata type . so we need to explecitly tell the datatype as float

```python
#-> Copies the shape of tensor 'a' i.e (2,3) and creates a new tensor of (2,3) but with all random value each time runned

torch.rand_like(a,dtype=torch.float32)
```

```python
#-> Copies the shape of tensor 'a' i.e (2,3) and creates a new tensor of (2,3) but with all random value each time with same values

torch.manual_seed(10)
torch.rand_like(a,dtype=torch.float32)
```

# Inplace **Operations**

```python
m = torch.randint(size=(2,3),low=1,high=5)
n = torch.randint(size=(2,3),low=1,high=5)
```

```python
print(m)
print(n)
```

```python
m+n # The results are not stored anywhere
```

```python
m.add_(n)  #Stores the result in tensor m
```

```python
m
```

```python
m.relu_() #Applies relu function and stores the result in tensor 'm' .. Here "_" represents inplace
```

# Copying a tensor

```python
o = torch.rand(2,3)
o
```

```python
# If we do the traditional approach that we do in python like this :
p=o
# This will create a new tensor 'p' referencing to the same memory of 'o'. So if we make change in p then the change will also be in o
p[0][0]=9
print(p)
print(o)
```

```python
id(o) ## Memory location of 'O'
```

```python
id(p) ## Memory location of 'P'
```

```python
# To copy a tensor , This create a new tensor 'q' in seperate memory location
q = o.clone()
```

```python
id(q)
```
